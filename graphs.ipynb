```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Rayons=[3,4,5]

Masse = [0,100,200,
        400,800,1600,
        3200,6400]
Masse = np.array(Masse) 
errMasse = 1

R=  [
    [7.5,7.5,10.0,
    10.0,12.5,15.0,
    0,22.5],
    [5.0,7.5,7.5,
    10.0,12.5,15.0,
    0,30.0],
    [5.0,10.0,10.0,
    12.5,15.0,17.5,
    22.5,30.0]]

V=  [
    [5.0,5.0,5.0,
    7.5,7.5,10.0,
    0,12.5], 
    [2.5,5.0,5.0,
    5.0,7.5,10.0,
    0,15.0],
    [2.5,5.0,5.0,
    7.5,10.0,10.0,
    12.5,15.0]
    ]
B=  [
    [5.0,5.0,5.0,
    7.5,10.0,10.0,
    0,15.0], 
    [5.0,5.0,7.5,
    7.5,10.0,12.5,
    0,17.5],
    [5.0,5.0,7.5,
    7.5,10.0,12.5,
    15.0,20.0]
    ]
```

```python
ToutLesSpheres = [R, V, B]
Noms = ["Rose", "Vert", "Bleu"]
YoungsGPa = np.array([0.000142, 0.001180, 0.000358])
YoungsMpa = 1e3*YoungsGPa 
def GraphMasseDeformation(X : list, nomGraph : str, nb ,loglog = True):
    fig, ax = plt.subplots()
    for i in range(len(Rayons)):
        ax.errorbar(X[i], Masse, xerr = 1.5, yerr = errMasse, fmt = "o", label = Rayons[i], alpha = 0.7)
    plt.legend(loc='upper left')
    plt.title(nomGraph + " Y =" + "%.3f" %YoungsMpa[nb] + "MPa")
    plt.xlabel("Deformation diametre en mm")
    plt.ylabel("Masse en gr")
    if loglog:
        ax.set_xscale('log')
        ax.set_yscale('log')
    plt.show()
    plt.close()
    
```

```python
for i in range(len(ToutLesSpheres)):
    GraphMasseDeformation(ToutLesSpheres[i], Noms[i], i) 
```
